```python
df_orig.columns = [col.replace(' ', '_').lower() for col in df_orig.columns]
```

```python
type_only = df_orig[['vehicle_id', 'vehicle_type']]
type_only['vehicle_type'].unique()
```

```python
type_only['vehicle_type'] = type_only['vehicle_type'].str.replace('SCHOOL BUS TYPE ', '', regex=False).str.replace(' <33,000 GVW', '', regex=False).str.replace(' <26,000 GVW', '', regex=False).str.replace(' ', '', regex=False)
```

```python
df_out = df_out.merge(type_only.rename(columns={'vehicle_id': 'id'}), on='id', how='left')
```

```python
df_out = df_out.rename(columns={'vehicle_type': 'type'})
```

```python
df_out.to_csv('./data/buses.csv', index=False)
```

```python
df_out.loc[df_out['type'] == 'BWC', 'range'] = 360
df_out.loc[df_out['type'] == 'C', 'range'] = 300
df_out.loc[df_out['type'] == 'B', 'range'] = 300
```
